```python
class Complexnumber:
  def __init__(self, real, imag):
    self.real = real
    self.imag = imag 
  def get_data(self):
    print(f'{self.real}+{self.imag}j')
m = Complexnumber(6,2)
m.get_data()
k = Complexnumber(9, 4)
k.hope = 11 #new atrbt specific for object
print(k.real,k.imag, k.hope)
print(m.hope)
```

```python
class Data:
  def __init__(self):
    self.__price = 600
  def value(self):
    print('price {}'.format(self.__price))
  
  def got(self, a):
    self.__price = a
    
p = Data()
p.value()
p.__price = 900
p.value()
p.got(1000)
p.value()
```

```python
#polymorphism

class Color:
  def green(self):
    print('Its green color')
  def red(self):
    print('Its red color')
class Get:
  def green(self):
    print('Its green color2')
  def red(self):
    print('Its red color2')
def gold(a):
  a.green()

m = Color()
k = Get()

gold(m)
gold(k)

```
