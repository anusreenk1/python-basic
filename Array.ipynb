```python
# Program to divide one array into two arrays
import array
arr = array.array('i', [5,6,9,1,2,4,3,7,8,10])
def split(a):
  print('here divides one array into two')
  print(f'array to split: {a}')
  half = len(a)//2
  print('first half: ', a[:half])
  print('second half: ', a[half:])

split(arr)
```

```python
# Python program to print the elements of an array present on even and odd position

def sort(a):
  even = []
  odd = []
  for i in range(len(a)):
    if i%2 == 0:
      even.append(a[i])
    else:
      odd.append(a[i])
  return even, odd
arr = array.array('i', [7,8,9,1,5,6,2])
sort(arr)
```
