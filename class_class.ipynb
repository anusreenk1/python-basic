```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def change(self):
    self.name = 'preeth'
    self.age = 20
    print(f'new name: {self.name}')
    print(f'new age : {self.age}')
class Employee(Person):

  def __init__(self, rolno, name, age):
    Person.__init__(self,name, age)
    self.rolno = rolno

  def info(self):
    print(f'info_name: {self.name}')
    print(f'info_age : {self.age}')
    print(f'info_rolno: {self.rolno}')
    print('wait its processing...')
    Person.change(self)

p = Person('ravi', 45)
p.change()

e = Employee(11, 'kripa', 30)
e.info()

    
```

```python
class Flower:
  def __init__(self, name, color):
    self.name = name
    self.color = color
    print('Flowers are awesome!!')
  def leaf(self):
    print('Leaves in green color')
    print('leaves are in good condition')
class Plant:
  def __init__(self, seed, height):
    self.seed = seed
    self.height = height
    print('pant a plant and be happy')
  def trunk(self):
    print('Tree trunks are good in condition')

class Rose(Flower,Plant):
  def __init__(self, name, color, seed, height):
    super().__init__(name, color) 
    super(Flower, self).__init__(seed,height)
    print(f'roses are in {self.color} color')
    

  def garden(self):
    print('its a beautiful garden')
    super().trunk()
    super().leaf()

r = Rose('nam', 'white', 'seedless', 90)
r.garden()
```

```python
class Operation:
  def __init__(self,m1, m2):
    self.m1 = m1
    self.m2 = m2
  def __add__(self,other):
    m1 = self.m1 + other.m1
    m2 = self.m2 + other.m2
    s3 = Operation(m1,m2)
    return s3





s1 = Operation(23,56)
s2 = Operation(78,30)
s3 = s1 + s2
s3.m1
```

```python
class Change:
  def __init__(self, m1, m2):
    self.m1 = m1
    self.m2 = m2
    

  def __add__(self, o2):
    m1 = self.m1 + o2.m1 
    m2 = self.m2 + o2.m2 
   
    o4 = Change(m1,m2)
    return o4

o1 = Change(1,2)
o2 = Change(4,5)
o4 = o1 + o2
o4.m1
```
